```python
age_summary = df['Age'].describe()
age_summary
```

```python
# Visual representation of Age Distribution
sns.histplot(df['Age'], kde=True, bins=30, color='crimson')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Billing_Amount_Distribution
sns.boxplot(data=df, y='Billing Amount', color='skyblue')
plt.title('Billing Amount Distribution')
plt.ylabel('Billing Amount')
plt.show()
```

```python
# Relationship between Age and Test Results
sns.boxplot(data=df, x='Test Results', y='Age', palette='muted')
plt.title('Age vs Test Results')
plt.xlabel('Test Results')
plt.ylabel('Age')
plt.show()
```

```python
# Test Results by Medical Condition
condition_vs_test = df.groupby('Medical Condition')['Test Results'].value_counts().unstack()
print("Test Results by Medical Condition:")
print(condition_vs_test)

# Visualization for Medication
sns.countplot(data=df, x='Test Results', hue='Medication', palette='Set2')
plt.title('Test Results by Medication')
plt.xlabel('Test Results')
plt.ylabel('Count')
plt.legend(title='Medication')
plt.show()
```

```python
# Room usage frequency
room_usage = df['Room Number'].value_counts().head(10)
print("Top 10 Rooms by Usage:")
print(room_usage)

# Visualization
room_usage.plot(kind='bar', color='teal')
plt.title('Top 10 Rooms by Usage')
plt.xlabel('Room Number')
plt.ylabel('Number of Admissions')
plt.show()
```

```python
# Admission rates by month
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Admission Month'] = df['Date of Admission'].dt.month
admission_monthly = df['Admission Month'].value_counts().sort_index()
print("Admissions by Month:")
print(admission_monthly)
```

```python
# Monthly Admission Trends
admission_monthly.plot(kind='bar', color='lightgreen')
plt.title('Admissions by Month')
plt.xlabel('Month')
plt.ylabel('Number of Admissions')
plt.show()
```
